```python
import rasterio
import geopandas as gpd
import pandas as pd
from rasterstats import zonal_stats
import os
from pathlib import Path
from google.colab import files

# ==============================
# 1. Define Paths
# ==============================
# Assuming TIFF files are in this Google Drive folder as seen in hk5GjY6jYAEi
tiff_folder = '/content/drive/MyDrive/Henry_bachelor_project/tiffs_shapefiles'
# Assuming shapefile is in this Google Drive folder as seen in 26a22a34
plants_shapefile_path = '/content/drive/MyDrive/Henry_bachelor_project/data/shapefile/kernza_plant_polygons_2025.shp'
output_csv_filename = 'all_bands_biomass_per_plant.csv'

# ==============================
# 2. Load Plant Polygons GeoDataFrame
# ==============================
if not os.path.exists(plants_shapefile_path):
    print(f"Error: Plant shapefile '{plants_shapefile_path}' not found.")
    print("Please ensure the shapefile is correctly placed in your Google Drive.")
    # Fallback if the shapefile was uploaded directly to /content/
    if os.path.exists('/content/kernza_plant_polygons_2025.shp'):
        plants_shapefile_path = '/content/kernza_plant_polygons_2025.shp'
        print(f"Using shapefile from /content/: {plants_shapefile_path}")
    else:
        raise FileNotFoundError(f"Plant shapefile not found at {plants_shapefile_path} or /content/")

plants = gpd.read_file(plants_shapefile_path)
initial_plants_crs = plants.crs
print(f"Loaded plant polygons with CRS: {initial_plants_crs}")

# ==============================
# 3. Identify TIFF Files for Processing
# ==============================
tiff_files = sorted(list(Path(tiff_folder).glob('*.tif')) + list(Path(tiff_folder).glob('*.tiff')))

if not tiff_files:
    print(f"No TIFF files found in '{tiff_folder}'. Please check the path and file extensions.")
    # Attempt to check if files were uploaded to /content directly
    content_tiffs = sorted(list(Path('/content/').glob('*.tif')) + list(Path('/content/').glob('*.tiff')))
    if content_tiffs:
        print(f"Found TIFF files in /content/: {', '.join([str(p.name) for p in content_tiffs])}")
        print("Consider moving them to Google Drive or updating 'tiff_folder' variable.")
    raise FileNotFoundError(f"No TIFF files found in {tiff_folder}")

print(f"Found {len(tiff_files)} TIFF files for processing:")
for f in tiff_files:
    print(f"- {f.name}")

# ==============================
# 4. Initialize Results Collection
# ==============================
all_stats_dfs = []

# ==============================
# 5. Process Each TIFF Image (Memory Efficient)
# ==============================
for i, tiff_path in enumerate(tiff_files):
    print(f"\nProcessing {i+1}/{len(tiff_files)}: {tiff_path.name}")
    try:
        with rasterio.open(str(tiff_path)) as src:
            raster_crs = src.crs
            print(f"  Raster CRS: {raster_crs}")

            # Reproject plants GeoDataFrame to match raster CRS if necessary
            if plants.crs != raster_crs:
                print(f"  Reprojecting plant polygons from {plants.crs} to {raster_crs}...")
                plants_reprojected = plants.to_crs(raster_crs)
            else:
                plants_reprojected = plants
                print("  Plant polygons already match raster CRS.")

            # Get filename for column prefixes
            file_prefix = Path(tiff_path).stem.replace('.', '_') # Replace dots for valid column names

            # Calculate zonal statistics
            stats = zonal_stats(
                plants_reprojected,
                str(tiff_path),
                stats=["mean", "median", "std"],
                nodata=None, # Assuming nodata is handled by rasterio internally or set to default if needed
                all_touched=True # Include polygons that only touch the raster
            )

            # Convert stats to DataFrame and rename columns
            current_stats_df = pd.DataFrame(stats)

            # Ensure columns are uniquely named for merging
            renamed_columns = {}
            for col in current_stats_df.columns:
                renamed_columns[col] = f"{file_prefix}_{col}"
            current_stats_df = current_stats_df.rename(columns=renamed_columns)

            all_stats_dfs.append(current_stats_df)

    except Exception as e:
        print(f"  Error processing {tiff_path.name}: {e}")
        print("  Skipping this TIFF file.")

# ==============================
# 6. Combine All Results
# ==============================
if not all_stats_dfs:
    print("No statistics were generated from any TIFF files. Cannot create combined CSV.")
else:
    print("\nCombining all zonal statistics...")
    # Concatenate all individual statistics DataFrames
    combined_stats_df = pd.concat(all_stats_dfs, axis=1)

    # Merge with the original plants GeoDataFrame (excluding geometry)
    # This assumes a stable order or implicit index alignment
    final_combined_df = pd.concat([
        plants.drop(columns='geometry').reset_index(drop=True),
        combined_stats_df.reset_index(drop=True)
    ], axis=1)

    print("Combined results head:")
    print(final_combined_df.head())

    # ==============================
    # 7. Save Combined Results to CSV
    # ==============================
    # Define output path in Google Drive
    output_drive_path = os.path.join(os.path.dirname(plants_shapefile_path), output_csv_filename)
    final_combined_df.to_csv(output_drive_path, index=False)
    print(f"\nSuccessfully saved combined results to '{output_drive_path}'")

    # ==============================
    # 8. Download Combined Results CSV
    # ==============================
    try:
        files.download(output_drive_path)
        print(f"Successfully downloaded '{output_csv_filename}'")
    except Exception as e:
        print(f"Could not automatically download '{output_csv_filename}': {e}")
        print("You can manually download it from Google Drive or the file browser in Colab.")

print("\nTask completed.")
```

```python
import rasterio
import numpy as np
import geopandas as gpd
import pandas as pd
from rasterstats import zonal_stats
import os
```

```python
BASE_PATH = "/content/drive/MyDrive/Biomass_Estimation_Project/01_RawData/Indices"
```

```python
RED_PATH     = os.path.join(BASE_PATH, "red.tif")
NIR_PATH     = os.path.join(BASE_PATH, "nir.tif")
REDEDGE_PATH = os.path.join(BASE_PATH, "red_edge.tif")
GREEN_PATH   = os.path.join(BASE_PATH, "green.tif")
SHAPEFILE    = os.path.join(BASE_PATH, "shapefile.shp")
```

```python
OUTPUT_PATH  = os.path.join(BASE_PATH, "Outputs")
os.makedirs(OUTPUT_PATH, exist_ok=True)
```

```python
required_files = {
    "red":      RED_PATH,
    "nir":      NIR_PATH,
    "red_edge":  REDEDGE_PATH,
    "green":    GREEN_PATH,
    "shapefile":SHAPEFILE
}

all_found = True
for name, path in required_files.items():
    if not os.path.exists(path):
        print(f"❌ MISSING [{name}]: '{path}'")
        all_found = False
    else:
        print(f"✅ Found  [{name}]: '{path}'")

if not all_found:
    raise FileNotFoundError("One or more files are missing. Check BASE_PATH and filenames.")
```

```python
def read_band(path):
    with rasterio.open(path) as src:
        arr = src.read(1).astype("float32")
        profile = src.profile
        crs = src.crs
    return arr, profile, crs
```

```python
print("\n📡 Computing vegetation indices...")

RED,   prof_red,  crs_ref = read_band(RED_PATH)
NIR,   _,         _       = read_band(NIR_PATH)
RE,    _,         _       = read_band(REDEDGE_PATH)
GREEN, _,         _       = read_band(GREEN_PATH)

np.seterr(divide="ignore", invalid="ignore")

NDVI  = np.where((NIR + RED)   != 0, (NIR - RED)   / (NIR + RED),             np.nan)
NDRE  = np.where((NIR + RE)    != 0, (NIR - RE)    / (NIR + RE),              np.nan)
GNDVI = np.where((NIR + GREEN) != 0, (NIR - GREEN) / (NIR + GREEN),           np.nan)
VARI  = np.where((GREEN + RED) != 0, (GREEN - RED) / (GREEN + RED),           np.nan)

print("✅ All indices computed.")
```

```python
print("\n💾 Saving index rasters to Drive...")

index_arrays = {
    "NDVI":  NDVI,
    "NDRE":  NDRE,
    "GNDVI": GNDVI,
    "VARI":  VARI,
}

output_profile = prof_red.copy()
output_profile.update(dtype="float32", count=1, nodata=np.nan)

saved_tiffs = {}
for index_name, array in index_arrays.items():
    out_path = os.path.join(OUTPUT_PATH, f"{index_name.lower()}_output.tif")
    with rasterio.open(out_path, "w", **output_profile) as dst:
        dst.write(array.astype("float32"), 1)
    saved_tiffs[index_name] = out_path
    print(f"  ✅ Saved: {out_path}")
```

```python
print("\n🗺️  Loading shapefile...")
plants = gpd.read_file(SHAPEFILE)
plants = plants.to_crs(crs_ref)
print(f"   CRS        : {crs_ref}")
print(f"   Polygons   : {len(plants)}")
```

```python
print("\n📊 Running zonal statistics...")

all_stats = {}
for index_name, tiff_path in saved_tiffs.items():
    stats = zonal_stats(
        plants, tiff_path,
        stats=["mean", "median", "std", "min", "max"],
        nodata=np.nan
    )
    stats_df = pd.DataFrame(stats)
    stats_df.columns = [
        f"{index_name}_mean", f"{index_name}_median",
        f"{index_name}_std",  f"{index_name}_min", f"{index_name}_max"
    ]
    all_stats[index_name] = stats_df
    print(f"  ✅ {index_name} done")
```

```python
attributes = plants.drop(columns="geometry").reset_index(drop=True)
combined   = pd.concat([attributes] + list(all_stats.values()), axis=1)

csv_path = os.path.join(OUTPUT_PATH, "vegetation_indices_per_plant.csv")
combined.to_csv(csv_path, index=False)

print(f"\n✅ ALL DONE!")
print(f"📁 Outputs saved to: {OUTPUT_PATH}")
print(f"📊 CSV : {csv_path}")
print(f"🗂️  TIFs: {list(saved_tiffs.values())}")
```

```python
import rasterio
import numpy as np
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from rasterstats import zonal_stats
from rasterio.warp import reproject, Resampling
from rasterio.transform import from_bounds
import os
```

```python
BASE_PATH   = "/content/drive/MyDrive/Biomass_Estimation_Project/01_RawData/Indices"

DSM_PATH    = os.path.join(BASE_PATH, "dsm.tif")
DTM_PATH    = os.path.join(BASE_PATH, "dtm.tif")
SHAPEFILE   = os.path.join(BASE_PATH, "shapefile.shp")

OUTPUT_PATH = os.path.join(BASE_PATH, "Outputs")
os.makedirs(OUTPUT_PATH, exist_ok=True)

CHM_PATH    = os.path.join(OUTPUT_PATH, "chm.tif")
VOL_PATH    = os.path.join(OUTPUT_PATH, "volume.tif")
```

```python
required = {"DSM": DSM_PATH, "dtm": DTM_PATH, "Shapefile": SHAPEFILE}

all_found = True
for name, path in required.items():
    if not os.path.exists(path):
        print(f"❌ MISSING [{name}]: '{path}'")
        all_found = False
    else:
        print(f"✅ Found  [{name}]: '{path}'")

if not all_found:
    raise FileNotFoundError("One or more files are missing. Check BASE_PATH and filenames.")
```

```python
print("\n🔄 Aligning DSM to DTM grid...")

with rasterio.open(DTM_PATH) as dtm_src:
    dtm_profile = dtm_src.profile
    dtm_crs     = dtm_src.crs
    dtm_trans   = dtm_src.transform
    dtm_shape   = (dtm_src.height, dtm_src.width)
    DTM         = dtm_src.read(1).astype("float32")
    DTM[DTM == dtm_src.nodata] = np.nan  # mask nodata

with rasterio.open(DSM_PATH) as dsm_src:
    DSM_aligned = np.empty(dtm_shape, dtype="float32")
    reproject(
        source      = rasterio.band(dsm_src, 1),
        destination = DSM_aligned,
        src_transform  = dsm_src.transform,
        src_crs        = dsm_src.crs,
        dst_transform  = dtm_trans,
        dst_crs        = dtm_crs,
        resampling     = Resampling.bilinear
    )
    DSM_aligned[DSM_aligned == dsm_src.nodata] = np.nan  # mask nodata

print("✅ Alignment done.")
```

```python
print("\n🌿 Computing CHM (DSM - DTM)...")

CHM = DSM_aligned - DTM

# Clip negative values to 0 (ground = 0, no underground values)
CHM = np.where(CHM < 0, 0, CHM)
CHM = np.where(np.isnan(DTM) | np.isnan(DSM_aligned), np.nan, CHM)

print(f"   CHM Min    : {np.nanmin(CHM):.3f} m")
print(f"   CHM Max    : {np.nanmax(CHM):.3f} m")
print(f"   CHM Mean   : {np.nanmean(CHM):.3f} m")
print("✅ CHM computed.")
```

```python
chm_profile = dtm_profile.copy()
chm_profile.update(dtype="float32", count=1, nodata=np.nan)

with rasterio.open(CHM_PATH, "w", **chm_profile) as dst:
    dst.write(CHM.astype("float32"), 1)

print(f"\n💾 CHM saved: {CHM_PATH}")
```

```python
print("\n📦 Computing volume per pixel...")

pixel_width  = abs(dtm_trans.a)   # metres per pixel (X)
pixel_height = abs(dtm_trans.e)   # metres per pixel (Y)
pixel_area   = pixel_width * pixel_height

print(f"   Pixel size : {pixel_width:.4f} m × {pixel_height:.4f} m")
print(f"   Pixel area : {pixel_area:.6f} m²")

VOLUME = np.where(np.isnan(CHM), np.nan, CHM * pixel_area)

print(f"   Volume Min : {np.nanmin(VOLUME):.4f} m³")
print(f"   Volume Max : {np.nanmax(VOLUME):.4f} m³")
print("✅ Volume computed.")
```

```python
vol_profile = dtm_profile.copy()
vol_profile.update(dtype="float32", count=1, nodata=np.nan)

with rasterio.open(VOL_PATH, "w", **vol_profile) as dst:
    dst.write(VOLUME.astype("float32"), 1)

print(f"\n💾 Volume raster saved: {VOL_PATH}")
```

```python
print("\n📊 Running zonal statistics per plant polygon...")

plants = gpd.read_file(SHAPEFILE).to_crs(dtm_crs)
print(f"   Polygons   : {len(plants)}")

chm_stats = zonal_stats(
    plants, CHM_PATH,
    stats=["mean", "max", "std", "min"],
    nodata=np.nan
)
vol_stats = zonal_stats(
    plants, VOL_PATH,
    stats=["sum", "mean", "max"],
    nodata=np.nan
)

chm_df = pd.DataFrame(chm_stats).rename(columns={
    "mean": "CHM_mean_m",
    "max":  "CHM_max_m",
    "min":  "CHM_min_m",
    "std":  "CHM_std_m"
})

vol_df = pd.DataFrame(vol_stats).rename(columns={
    "sum":  "Volume_total_m3",
    "mean": "Volume_mean_m3",
    "max":  "Volume_max_m3"
})
```

```python
attributes = plants.drop(columns="geometry").reset_index(drop=True)
final_df   = pd.concat([attributes, chm_df, vol_df], axis=1)

csv_path = os.path.join(OUTPUT_PATH, "chm_volume_per_plant.csv")
final_df.to_csv(csv_path, index=False)

print(f"\n📋 Preview:")
print(final_df.head())
print(f"\n💾 CSV saved: {csv_path}")
```
